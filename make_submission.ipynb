```python
import os
import glob
import h5py
import numpy as np
from keras import backend as K
from keras.models import Sequential, load_model
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.layers.core import Flatten, Dense, Activation, Reshape
from keras.layers import Dropout
from keras.preprocessing.image import ImageDataGenerator
from keras import optimizers
from keras.callbacks import Callback
import cv2
```

```python
print(K.image_dim_ordering())
print(K.image_data_format())
img_width, img_height = 150, 150
root_train_folder_path = os.path.join('database')
train_data_dir = os.path.join(root_train_folder_path, 'train')
validation_data_dir = os.path.join(root_train_folder_path, 'validation')
```

```python
# used to rescale the pixel values from [0, 255] to [0, 1] interval
datagen = ImageDataGenerator(rescale=1./255)
```

```python
# automagically retrieve images and their classes for train and validation sets
train_generator = datagen.flow_from_directory(
        train_data_dir,
        target_size=(img_width, img_height),
        batch_size=16,
        class_mode='binary')

validation_generator = datagen.flow_from_directory(
        validation_data_dir,
        target_size=(img_width, img_height),
        batch_size=4,
        class_mode='binary')
```

```python
models_folder_path = os.path.join("models")
model_name = "continue_callback_model"
model = load_model(os.path.join(models_folder_path, model_name) + '-epoch{}.h5'.format(45))
```

```python
model.summary()
```

```python
def img_preprocessing(img_path):
    np_img = cv2.imread(img_path)
    # print(np_img)
    np_img = cv2.resize(np_img, (img_width, img_height))
    np_img = np_img / 255.0
    np_img = np.expand_dims(np_img, axis=0)
    return np_img
```

```python
file_name_list = []
result_list = []

for each_test_file in os.listdir('test'):
    batch_img = img_preprocessing(os.path.join("test", each_test_file))
    file_name_list.append(each_test_file)
    ret = model.predict_classes(batch_img)
    result_list.append(ret[0][0])

print(len(file_name_list))
print(len(result_list))
```
